## Final Analysis: Country Development Index

#### The objective of this analysis is to combine key indicators (like GDP, Life Expectancy, Unemployment Rate, etc.) into a composite Country Development Index (CDI) that reflects each country’s overall development level.

#### Load and normalize the data. 

#### To combine different indicators, we have to scale them to a common range. Hence, we'll be normalizing each key indicator on a scale of 0 to 1. The scale essentially assigns weights to each indicator based on its perceived importance in assessing development. 

#### For example:


* #### GDP per capita: 0.3
* #### Life Expectancy: 0.3
* #### Literacy Rate: 0.2
* #### Unemployment Rate: 0.1
* #### Infant Mortality Rate:: 0.1)

```python
USA_vs_UK3 = data[(data['Country Name'].isin(['United States', 'United Kingdom']))]
```

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())
#USA_vs_Nigeria['InfMort_normalized'] = 1 - normalize(USA_vs_Nigeria['Infant mortality rate (per 1,000 live births)'])   # Inverse since lower infant mortality is better


USA_vs_UK3.loc[:, 'GDP_normalized'] = normalize(USA_vs_UK3['GDP per capita (USD)'])
USA_vs_UK3.loc[:, 'LifeExp_normalized'] = normalize(USA_vs_UK3['Life expectancy at birth (years)'])
USA_vs_UK3.loc[:, 'EleectricConsum_normalized'] = normalize(USA_vs_UK3['Electric power consumption (kWh per capita)'])
USA_vs_UK3.loc[:, 'Unemployment_normalized'] = 1 - normalize(USA_vs_UK3['Unemployment (% of total labor force) (modeled ILO estimate)'])
USA_vs_UK3.loc[:, 'InfMort_normalized'] = 1 - normalize(USA_vs_UK3['Infant mortality rate (per 1,000 live births)'])
```

#### Next, we compute the Development Index Score: MMultiply each normalized indicator by its weight and sum them to get the CDI for each country.

```python
USA_vs_UK3.loc[:, 'CDI'] = (
    0.3 * USA_vs_UK3.loc[:, 'GDP_normalized'] +
    0.3 * USA_vs_UK3.loc[:, 'LifeExp_normalized'] +
    0.2 * USA_vs_UK3.loc[:, 'EleectricConsum_normalized'] +
    0.1 * USA_vs_UK3.loc[:, 'Unemployment_normalized'] +
    0.1 * USA_vs_UK3.loc[:, 'InfMort_normalized']
)

#Rescale the CDI: If you want the CDI on a scale of 0 to 100, multiply it by 100.

USA_vs_UK3.loc[:, 'CDI'] = USA_vs_UK3.loc[:, 'CDI'] * 100
```

#### Then, sort and plot the CDI values to rank countries by development level..


```python
numeric_data = USA_vs_UK3.select_dtypes(include='number')

data_sorted = numeric_data.groupby(USA_vs_UK3['Country Name']).mean().sort_values(by='CDI', ascending=False)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.barh(data_sorted.index, data_sorted['CDI'], color='skyblue')
plt.xlabel('Country Development Index')
plt.ylabel('Country')
plt.title('Country Development Index by Country')
plt.show()
```

#### The chart shows a slightly higher Country Development Index for the United States, compared to the United Kingdom. This would imply the US has a slightly better standard of living and overall well-being

#### That said, the Development Index is measured by each country's GDP, Life Expectancy, Electricity Consumption, Infant Mortality Rate, and Percentage Unemployment. Considering other economic, social, political, and environmental developments might skew the result.

#### Finally, we can perform a Time Series Analysis: Here, we plot the CDI over time for both countries to observe development trends.

```python
import seaborn as sns

sns.lineplot(x='Year', y='CDI', hue='Country Name', data=USA_vs_UK3)
plt.title('Country Development Index Over Time')
plt.xlabel('Year')
plt.ylabel('CDI')
plt.show()
```

#### From the time graph, both countries show an upward trend in their CDI overtime, indicating general development progress. However, the United States have maintained a higher level, meaning it has generally outperformed the United Kingdom regarding development.

#### That said, the United Kingdom has taken a steep increase in CDI, and if this trend continues, the United Kingdom will be at par with the United States. We can see it from the next visualization.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


# Assuming 'Country Name' is the correct column; replace with the actual column if needed
country_column = 'Country Name'

# Copy the DataFrame to avoid modifying the original data
USA_vs_UK3 = USA_vs_UK3.copy()

# Drop rows with NaN in 'CDI' to avoid errors in fitting
USA_vs_UK3 = USA_vs_UK3.dropna(subset=['CDI'])

# Prepare for storing predictions
future_predictions = []

# Loop through each country to create predictions
for country in USA_vs_UK3[country_column].unique():
    # Filter data for the current country
    country_data = USA_vs_UK3[USA_vs_UK3[country_column] == country]
    
    # Define the feature (Year) and target (CDI) for the linear model
    X = country_data['Year'].values.reshape(-1, 1)
    y = country_data['CDI'].values
    
    # Fit a linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Generate years for the next 30 years and predict CDI
    last_year = country_data['Year'].max()
    future_years = np.arange(last_year + 1, last_year + 31).reshape(-1, 1)
    future_cdi = model.predict(future_years)
    
    # Create a DataFrame for the future predictions
    future_data = pd.DataFrame({
        'Year': future_years.flatten(),
        'CDI': future_cdi,
        country_column: country
    })
    future_predictions.append(future_data)

# Combine the original data with the future predictions
USA_vs_UK_with_predictions = pd.concat([USA_vs_UK3] + future_predictions)

# Plotting the data including future predictions
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='CDI', hue=country_column, data=USA_vs_UK_with_predictions)
plt.title('Country Development Index Over Time with 30-Year Future Predictions')
plt.xlabel('Year')
plt.ylabel('CDI')
plt.show()
```
